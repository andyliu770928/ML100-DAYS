# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = './data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
# 先篩選數值型的欄位
"""
YOUR CODE HERE, fill correct data types (for example str, float, int, ...)
"""
# 此處選擇float與int是為了計算Outlier
dtype_select = [pd.np.dtype('float64'),pd.np.dtype('int64')]
numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))])
print(len(numeric_columns))
# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns2 = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2.0 ))])
print("Numbers of remain %s columns" % len(numeric_columns2))
# 把有缺失值的row刪除


# 檢視這些欄位的數值範圍
outlier_col=[]
for col in numeric_columns2:
    """
    Your CODE HERE, make the box plot
    """
    # 1st quartile (25%)
    Q1 = np.nanpercentile(app_train[col],25)
    # 3rd quartile (75%)
    Q3 = np.nanpercentile(app_train[col],75)
    # Interquartile range (IQR)
    IQR = Q3 - Q1
    # outlier step
    outlier_step = 3.0 * IQR 
    
    if (app_train[col].max() > (Q3+outlier_step)) or (app_train[col].min() < (Q1-outlier_step)):
        plt.boxplot(app_train[col].dropna(),whis=3)
        plt.title(col)
        plt.show()
        outlier_col.append(col)
```

```python
# 以下都是有Outlier的特徵(超過三倍IQR)
print(outlier_col)
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""

series = pd.Series(app_train['AMT_INCOME_TOTAL'])
cdf = series.value_counts().sort_index().cumsum()

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```
